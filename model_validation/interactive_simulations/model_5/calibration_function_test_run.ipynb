```python
from datetime import datetime, timedelta
from pathlib import Path
import itertools
from vivarium import Artifact, InteractiveContext
import ipywidgets
import pandas as pd, numpy as np, os
import sys


def run_cohort_simulation(sim_config,
                          input_draw,
                          random_seed,
                          x_factor_effect_to_mam,
                          x_factor_effect_to_sam,
                          x_factor_exposure
                          ):
    cohort_baseline = pd.DataFrame()
    cohort_endline = pd.DataFrame()
    controls_endline = pd.DataFrame()
    population_size = 10000
    sim = InteractiveContext(Path(sim_config), setup=False)
    sim.configuration.update({'input_data':
                                  {'input_draw_number': input_draw},
                              'randomness':
                                  {'random_seed': random_seed},
                              'population':
                                  {'population_size': population_size},
                              #'effect_of_x_factor_on_mild_child_wasting':
                              #    {'incidence_rate':
                              #         {'relative_risk': x_factor_effect_to_mild}},
                              'effect_of_x_factor_on_mild_child_wasting_to_moderate_acute_malnutrition':
                                  {'transition_rate':
                                       {'relative_risk': x_factor_effect_to_mam}},
                              'effect_of_x_factor_on_moderate_acute_malnutrition_to_severe_acute_malnutrition':
                                  {'transition_rate':
                                       {'relative_risk': x_factor_effect_to_sam}},
                              #'child_wasting':
                              #    {
                              #        'mild_child_wasting_untreated_recovery_time': mild_child_wasting_untreated_recovery_time},
                              'x_factor':
                                  {'exposure': x_factor_exposure}
                              })
    sim.setup()

    cohort_start_recruiting_time = pd.Timestamp(year=2022, month=1, day=2)
    cohort_recruitment_duration_in_timesteps = 183
    observation_duration_in_days = 365

    #cohort_start_recruiting_time = pd.Timestamp(year=2021, month=7, day=2)
    #cohort_recruitment_duration_in_timesteps = 2
    #observation_duration_in_days = 1

    sim.run_until(cohort_start_recruiting_time)
    pop_t = sim.get_population()

    controls_baseline = pop_t.loc[(pop_t.child_wasting.isin(['susceptible_to_child_wasting',
                                                             'mild_child_wasting']))
                                  & (pop_t.age > 0.5)]

    for step in list(range(0, cohort_recruitment_duration_in_timesteps + observation_duration_in_days * 2)):

        sim.step()
        pop_t_minus_1 = pop_t.copy()
        pop_t = sim.get_population()
        pop_t['status'] = np.nan

        if step <= cohort_recruitment_duration_in_timesteps:
            # add simulants to the cohort_baseline dataframe upon recovery from SAM or MAM
            # note nature of their recovery as "status"
            # record date of recovery so that they can be followed for a fixed time post-recovery
            for i in [i for i in pop_t.index
                      if i in pop_t_minus_1.index
                         and i not in cohort_baseline.index]:

                if ((pop_t_minus_1['child_wasting'].loc[i] == 'moderate_acute_malnutrition')
                        & (pop_t['child_wasting'].loc[i] == 'mild_child_wasting')):

                    pop_t.loc[i, 'status'] = 'recovered_from_mam'
                    cohort_baseline = cohort_baseline.append(
                        pd.DataFrame(pop_t.loc[i]).transpose().rename(columns={'mild_child_wasting_event_time':
                                                                                   'cohort_baseline_entrance_time'}))

                elif pop_t_minus_1['child_wasting'].loc[i] == 'severe_acute_malnutrition':

                    if pop_t['child_wasting'].loc[i] == 'mild_child_wasting':
                        pop_t.loc[i, 'status'] = 'recovered_from_sam'
                        cohort_baseline = cohort_baseline.append(
                            pd.DataFrame(pop_t.loc[i]).transpose().rename(columns={'mild_child_wasting_event_time':
                                                                                       'cohort_baseline_entrance_time'}))
                    elif pop_t['child_wasting'].loc[i] == 'moderate_acute_malnutrition':
                        pop_t.loc[i, 'status'] = 'defaulted_from_sam_to_mam'
                        cohort_baseline = cohort_baseline.append(
                            pd.DataFrame(pop_t.loc[i]).transpose().rename(
                                columns={'moderate_acute_malnutrition_event_time':
                                             'cohort_baseline_entrance_time'}))
        # if it has been one year past entering cohort and they are still <5 years of age (tracked), add them to the endline data frame
        # record their x-factor and wasting treatment exposures as well as MAM/SAM event counts at one year
        if step >= observation_duration_in_days * 2:
            for i in [i for i in cohort_baseline.index if i in pop_t.index and i not in cohort_endline.index]:

                if ((sim._clock.time - cohort_baseline.loc[i, 'cohort_baseline_entrance_time'])
                        > np.timedelta64(observation_duration_in_days, 'D')):
                    cohort_endline = cohort_endline.append(pd.DataFrame(pop_t.loc[i]).transpose())
                    #cohort_endline.loc[i, 'x_factor_exposure'] = sim.get_value('x_factor.exposure')(pop_t.index).loc[i]
                    #cohort_endline.loc[i, 'wasting_treatment_exposure'] = sim.get_value('wasting_treatment.exposure')(pop_t.index).loc[i]

        # record data for healthy controls one year after baseline data
        if step == observation_duration_in_days * 2:
            controls_endline = pop_t.loc[[i for i in controls_baseline.index if i in pop_t.index]]
            #controls_endline['x_factor_exposure'] = sim.get_value('x_factor.exposure')(pop_t.index).loc[
            #    controls_endline.index]
            #controls_endline['wasting_treatment_exposure'] = sim.get_value('wasting_treatment.exposure')(pop_t.index).loc[controls_endline.index]

    return cohort_baseline, cohort_endline, controls_baseline, controls_endline


def format_and_save_results(outputdir,
                            cohort_baseline,
                            cohort_endline,
                            controls_baseline,
                            controls_endline,
                            input_draw,
                            random_seed,
                            x_factor_effect_to_mam,
                            x_factor_effect_to_sam,
                            x_factor_exposure):
    cohort_baseline['timepoint'] = 'baseline'
    cohort_endline['timepoint'] = 'endline'
    cohort = pd.concat([cohort_baseline, cohort_endline])
    cohort['population'] = 'cohort'

    controls_baseline['timepoint'] = 'baseline'
    controls_endline['timepoint'] = 'endline'
    controls = pd.concat([controls_baseline, controls_endline])
    controls['population'] = 'controls'

    data = pd.concat([cohort, controls])
    data = data[['alive', 'entrance_time', 'age', 'sex', 'child_wasting',
                 'susceptible_to_child_wasting_event_count',
                 'susceptible_to_child_wasting_event_time',
                 'mild_child_wasting_event_count', 'cohort_baseline_entrance_time',
                 'moderate_acute_malnutrition_event_count',
                 'moderate_acute_malnutrition_event_time',
                 'severe_acute_malnutrition_event_time',
                 'severe_acute_malnutrition_event_count', 'status',
                 'mild_child_wasting_event_time', 'timepoint', #'x_factor_exposure',
                 #'wasting_treatment_exposure',
                 'population']]
    data['input_draw'] = input_draw
    data['random_seed'] = random_seed
    data['x_factor_effect_to_mam'] = x_factor_effect_to_mam
    data['x_factor_effect_to_sam'] = x_factor_effect_to_sam
    data['x_factor_exposure'] = x_factor_exposure

    subdir = f"draw_{input_draw}_seed_{random_seed}_x_factor_effect_{x_factor_effect_to_mam}_{x_factor_effect_to_sam}_x_factor_exposure_{x_factor_exposure}"
    os.makedirs(outputdir + subdir)
    data.to_csv(outputdir + subdir + '/data.csv')


def main_function(sim_config,
                  outputdir,
                  input_draw,
                  random_seed,
                  x_factor_effect_to_mam,
                  x_factor_effect_to_sam,
                  x_factor_exposure
                  ):
    (cohort_baseline,
     cohort_endline,
     controls_baseline,
     controls_endline) = run_cohort_simulation(sim_config,
                                               input_draw,
                                               random_seed,
                                               x_factor_effect_to_mam,
                                               x_factor_effect_to_sam,
                                               x_factor_exposure)
    format_and_save_results(outputdir,
                            cohort_baseline, cohort_endline, controls_baseline, controls_endline,
                            input_draw, random_seed, x_factor_effect_to_mam,
                            x_factor_effect_to_sam,
                            x_factor_exposure)

```

```python
cols = ['input_draw','random_seed',
        'x_factor_effect_to_mam','x_factor_effect_to_sam',
        'x_factor_exposure',
        #'run',
        'simulant']

def calculate_episode_frequency_stratified(status, severity, strata_cols):
    if status=='control':
        baseline = (data.loc[(data.population=='controls') & (data.timepoint=='baseline')]
                .set_index(cols))
        endline = (data.loc[(data.population=='controls') & (data.timepoint=='endline')]
               .set_index(cols))
    else:
        baseline = (data.loc[data.status==status]
                    .set_index(cols).drop(columns=strata_cols))
        endline = (data.loc[(data.population=='cohort') & (data.timepoint=='endline')]
                   .set_index(cols))
    res = baseline.merge(endline, right_index=True, left_index=True, suffixes=['_baseline','_endline']).reset_index()
    res[f'{severity}_episodes'] = res[f'{severity}_event_count_endline'] - res[f'{severity}_event_count_baseline']
    res['count'] = 1
    res = (res.groupby([c for c in cols if c != 'random_seed' and c != 'simulant'] 
                       + strata_cols + [f'{severity}_episodes']).sum()
              / res.groupby([c for c in cols if c != 'random_seed' and c != 'simulant'] + strata_cols).sum())[['count']].reset_index()
    res = res.loc[res[f'{severity}_episodes']==0]
    res[f'{status}_{severity}_relapse_frequency'] = 1 - res['count']   
    
    return res.set_index([c for c in cols if 
                          c != 'simulant' and c != 'random_seed'] 
                         + strata_cols)[[f'{status}_{severity}_relapse_frequency']]
```

```python
def calculate_relapse_rates(strata_cols):
    sam_sam = calculate_episode_frequency_stratified('recovered_from_sam', 'severe_acute_malnutrition', strata_cols)
    sam_mam = calculate_episode_frequency_stratified('recovered_from_sam', 'moderate_acute_malnutrition', strata_cols)
    #mam_sam = calculate_episode_frequency_stratified('recovered_from_mam', 'severe_acute_malnutrition', strata_cols)
    mam_mam = calculate_episode_frequency_stratified('recovered_from_mam', 'moderate_acute_malnutrition', strata_cols)
    controls_sam = calculate_episode_frequency_stratified('control', 'severe_acute_malnutrition', strata_cols)
    sam = sam_sam.merge(sam_mam, right_index=True, left_index=True)
    #mam = mam_sam.merge(mam_mam, right_index=True, left_index=True)
    relapse = sam.merge(mam_mam, right_index=True, left_index=True)
    relapse = relapse.merge(controls_sam, right_index=True, left_index=True)
    return relapse
```
