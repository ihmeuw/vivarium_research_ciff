```python
from get_draws.api import get_draws
from db_queries import get_population
```

```python
df = get_draws(gbd_id_type='rei_id',
                 gbd_id=240,
                 source='exposure',
                 year_id=2020,
                 gbd_round_id=7,
                 status='best',
                 location_id = [179],
                 decomp_step = 'iterative')
df
```

```python
# filter to the age group IDs needed
age_list = [2,3,34,238,388,389]
df = df.query("age_group_id in @age_list")
```

```python
# create 2 dataframes that are filtered to the respective parameters
cat1_df = df.query("parameter == 'cat1'").reset_index(drop=True)
cat2_df = df.query("parameter == 'cat2'").reset_index(drop=True)
cat2_df
```

```python
# drop the 'parameter' column for each dataframe
cat1_df.drop("parameter", inplace=True, axis=1)
cat2_df.drop("parameter", inplace=True, axis=1)
cat1_df.drop("measure_id", inplace=True, axis =1)
cat2_df.drop("measure_id", inplace=True, axis =1)
cat1_df.drop("metric_id", inplace=True, axis =1)
cat2_df.drop("metric_id", inplace=True, axis =1)
```

```python
# set the index to all non-draw columns, so that we can add the two dataframes together.
index_list = ["age_group_id", "sex_id", "location_id", "year_id"]
cat1_op_df = cat1_df.set_index(index_list)
cat2_op_df = cat2_df.set_index(index_list)
```

```python
# add the two operands together
sum_op_df = cat1_op_df + cat2_op_df
sum_op_df
```

```python
# get the population dataframe and set its index so we can multiply with sum.
pop_df = get_population(age_group_id=[235, 2, 3, 4, 5], location_id=179, year_id=2020, gbd_round_id=7, decomp_step = 'iterative')
pop_df.run_id.unique()
pop_df.drop('run_id', inplace = True, axis = 1)

# creating draw_0 through draw_1000 with same population estimate for each value:
for i in range(1000):
     pop_df[f"draw_{i}"] = pop_df["population"]
pop_df.drop("population", axis=1, inplace=True)
pop_df = pop_df.reset_index(drop=True)
pop_op_df = pop_df.set_index(index_list)
pop_op_df
```

```python
# multiplying sums and populations
nums_df = (sum_op_df * pop_op_df).reset_index()
nums_df
```

```python
# sum over age -- group-by all columns except 'age_group_id' and 'draw_x'
group_col_list = ["sex_id", "location_id", "year_id"]
age_sum_nums_df = nums_df.groupby(group_col_list).sum().reset_index().drop("age_group_id", axis=1)
age_sum_pop_df = pop_df.groupby(group_col_list).sum().reset_index().drop("age_group_id", axis=1)
```

```python
# convert back to prevalence rate-space by dividing by aggregate population
age_sum_df = (age_sum_nums_df.set_index(group_col_list) / age_sum_pop_df.set_index(group_col_list)).reset_index()
age_sum_df
```
